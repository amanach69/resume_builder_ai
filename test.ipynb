```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.survived.value_counts().plot(kind='pie', figsize=(7,5), autopct='%1.1f%%', startangle=90)
plt.xlabel('Survival')
plt.title('Survival Rate')
plt.show()
```
